## Creative non-fiction

```python
# Load job data
# df_all = pd.read_csv('../synthesis/data.jobcensus.synthesized.2021-01-12.csv')
date='2021-05-05'

df_all = pd.read_csv(f'../synthesis/data.jobcensus.synthesized.{date}.csv')
# Load field aliases
df_alias = pd.read_csv('../synthesis/data.jobcensus.field_aliases.csv')
# Merge
df = df_all.merge(df_alias,on='JobField',how='inner')
# Clean
df['JobYear']=df['JobYear'].apply(int)
df = df[df.FieldAlias!='--']
# Show
df
```

```python
# Creative writing?
df['FieldAlias'].value_counts()
```

```python
df_cw = df[df['FieldAlias'] == "Creative Writing"]
df_cw
```

```python
# Mark as creative non/fiction?
def mark_creative_nonfic(jobdesc):
    import re
    # lowercase
    jobdesc=str(jobdesc).lower().replace('non-fiction','nonfiction')
    rgx = re.compile("(\w[\w']*\w|\w)")
    jobdesc_words=set(rgx.findall(jobdesc))
    labeld = {
        'Non-Fiction':{'non-fiction','nonfiction'},
        'Fiction':{'fiction'},
        'Poetry':{'poetry'}
    }
    label = []
    for genre,lbls in labeld.items():
        if lbls & jobdesc_words:
            label.append(genre)
    labelstr='/'.join(label)
    if not labelstr: labelstr='Other'
    return labelstr
```

```python
df_cw['cw_genre']=df_cw['JobDesc'].apply(mark_creative_nonfic)
df_cw['cw_genre_nonfiction']=df_cw['cw_genre'].apply(lambda x: 'Non-Fiction' if 'Non-Fiction' in x else 'Other')
df_cw['cw_genre_fiction']=df_cw['cw_genre'].apply(lambda x: 'Fiction' if 'Fiction' in x.replace('Non-Fiction','') else 'Other')
df_cw['cw_genre_poetry']=df_cw['cw_genre'].apply(lambda x: 'Poetry' if 'Poetry' in x else 'Other')
df_cw['cw_genre'].value_counts()
```

```python
# save
df_cw.to_csv(f'../synthesis/data.jobcensus.synthesized.{date}.creative-writing.csv')
```

```python
## Plot
def do_plot(color_by, save_to, msg=''):
    # get date
    now=datetime.now()
    date=f'{now.year}-{now.month:02d}-{now.day:02d}'
    
    # count
    df_count=df_cw.groupby([color_by,'JobYear']).count().reset_index()
    
    # make fig
    fig=(
        p9.ggplot(p9.aes(x='JobYear',y='Count',color=color_by), df_count) 
        + p9.geom_line()
        + p9.geom_text(p9.aes(label='Count'),size=10)
        + p9.scale_x_continuous(breaks=range(2011,2021))
        + p9.ggtitle(f'Number of creative writing jobs{msg}, 2011-2020 (as of {date})')
    )
    # save
    fig.save(save_to.replace('.png',f'.{date}.png'),dpi=300)

    #return
    return fig
```

```python
do_plot(
    'cw_genre',
    f'figures/figmisc1.creative_writing_jobs.{date}.png',
    ' by mentions of genre'
)
```

```python
do_plot(
    'cw_genre_nonfiction',
    f'figures/figmisc1.creative_writing_jobs.nonfic.{date}.png',
    ' in creative non-fiction'
)
```

```python
do_plot(
    'cw_genre_fiction',
    f'figures/figmisc1.creative_writing_jobs.fiction.{date}.png',
    ' in fiction'
)
```
